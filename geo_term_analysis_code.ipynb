```python
import requests
import re
import pandas as pd
import numpy

#this part of the code screens the spreadsheet with the sensitive mutants, and adds the contents of each cell
#to the list as long as it has LMJ as the first three characters
#make sure that the spreadsheet you are using is titled "sensitive_mutants.csv" as written in my code
#or replace "sensitive_mutants.csv" with whatever the name of your file is, in quotes, and make sure it is in csv format

excel = pd.read_csv("sensitive_mutants.csv", header = None)
x = [i for i in excel.values]
lmjlist = []
for i in x:
    for z in i:
        if z[:3] == "LMJ":
            lmjlist.append(z)        

            
#this function takes the list of LMJ codes as the input, and outputs a list of TAIR codes
#not all LMJ codes had a corresponding TAIR code, so the output is smaller than the input

def geo_term_generator(lmjlist):
    tair_list = []
    for i in lmjlist:
        sentinel = requests.get("https://www.chlamylibrary.org/showMutant?mutantName=" + i)
        txt = sentinel.text
        mypattern = 'geneIdentifier=Cre\d{2}\..{7}'
        my_re = re.compile(mypattern)    
        match_result = my_re.search(txt)
        if match_result:
            cre_name = match_result.group(0)
        else:
            continue
        sentinel2 = requests.get("https://www.chlamylibrary.org/showGene?" + cre_name)
        txt2 = sentinel2.text
        mypattern2 = 'type=gene&amp;name=AT.{7}\..'
        my_re2 = re.compile(mypattern2)
        match_result2 = my_re2.search(txt2)
        if match_result2:
            tair_name = match_result2.group(0)
        else:
            continue
        tair_list.append(tair_name.split("=")[-1])
    return tair_list

tair_list = geo_term_generator(lmjlist)
for z in tair_list:
    print(z)
```
